```python
# ✅ Cargar archivo info_capitulos.json de ejemplo
import json
from pathlib import Path

info_path = Path("libros/angelo_artefactos/info_capitulos.json")
with info_path.open(encoding='utf-8') as f:
    info = json.load(f)

cap_1 = info[0]
print(f"Primer capítulo: {cap_1['capitulo']}")
print(f"Ruta carpeta: {cap_1['carpeta']}")
print(f"Escenas: {cap_1['escenas'][:2]}...")
```

```python
# 📄 Leer contenido de una escena de ejemplo
ruta_escena = Path('libros/angelo_artefactos') / cap_1['escenas'][0].replace('escenas\\', 'escenas/')
with ruta_escena.open(encoding='utf-8') as f:
    contenido = f.read()
print(contenido[:500])
```
